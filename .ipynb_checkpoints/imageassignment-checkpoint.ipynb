```python
import os, os.path
import IPython.display
#import PIL as pil
from PIL import Image
import numpy as np

def main():
    try:
        dirname = "photos"
        count=0
        onlyfiles = next(os.walk(dirname))[2]
        mArray=np.zeros([len(onlyfiles),200,200,3],dtype=np.uint8)  # creating numpy array
    
        for filename in onlyfiles:
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                image = Image.open(dirname+'/'+filename)
                img_resized = image.resize((200,200))  # Resize Image using PIL Library
                #print(img_resized)
                img=np.array(img_resized)  # convert resize image to numpy array 
                mArray[count]=img
                # print("mArray image shape on position of array",count,mArray[count].shape)
                #print(Image.fromarray(mArray[count], 'RGB'))
                i=Image.fromarray(mArray[count], 'RGB')
                IPython.display.display(i)
                print(mArray[count].shape)
            count=count+1
        return mArray;

    except FileNotFoundError:
        print("File Found");
    except Exception as e:
        print("Some thing went wrong"+str(e))

if __name__ =="__main__":
    main()
```
